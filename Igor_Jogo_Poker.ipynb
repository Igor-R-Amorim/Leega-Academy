# Lógica do programa

Iniciar o menu com as opçoes: Sair, Regras e Iniciar Jogo

-   Caso o usuario escolha sair, finalizar o programa. 

-   Caso seja escolhido regras, mostras as regras e voltar ao menu principal.

-   Caso seja escolhido iniciar o jogo, apresentar o oponente e perguntar se o jogador que continuar o jogo ou sair.

 - Se o jogador escolher sair, voltar ao menu principal.

 - Se o jogador escolher continuar, dar inicio as Rodadas.

  - - A cada Rodada será aquirido dois novos baralhos. Um baralho de cor de fundo vermelho e um de cor de fundo azul.

  - - Esses baralhos serão embaralhados e distribuidos aleatoriamente para os jogadores sem saber a quantidade de cartas de cada baralho.
  - - Cada jogador receberá em sua mão 7 cartas do Dealer, que serão distribuidas do baralho azul e do baralho vermelho de forma aleatoria.
  - - O jogador que receber as suas 7 cartas deve 
escolher 2 cartas para eliminar da sua mão e 
adicionar ao morto, assim que terminar de eliminar
suas cartas o proximo jogador pode iniciar a fase dele.
 - - O proximo jogador deverá seguir os mesmo passos, escolher entre as 7 cartas da sua mão 2 para serem eliminadas.
 - - Apos todos os jogadores terem escolhidos suas cartas, ambos devem abaixar suas mãos para que seja contado quem fez a maior jogada.
 - - O programa deve ler carta por carta de cada mão para definir quem fez a maior jogada.
 - - Em caso de empate sera escolhido declarado o vencedor aquele que tiver a carta de maior valor em sua mao.
 - - caso os dois jogadores empatem com o mesmo valor de carta, sera escolhido o jogar que tiver o naipe mais forte. sendo a sequencia 
 
        (espadas) > (copas) > (ouros) > (paus)
 - - Se ambos tiverem as cartas de valor igual e de mesmo naipe, sera entao avaliado qual carta de valor mais alto e naipe mais forte, que possui a cor de fundo do baralho vermelha.
 - - Se ambas as maos forem identicas, entao deverá ser sacado uma carta para cada jogador seguindo as mesmas regras de desempate já descritas [ Maior valor > maior naipe > cor de fundo vermelha ]
  - - Em caso de continuar empatado devera ser sacado outra carta ate satisfazer alguma das condiçoes.
  - - apos definir o vencedor, deve-se contar quantas vezes cada jogador ganhou e mostrar a todos o placar de vitorias de cada um. 
  - - O jogo finaliza em 5 rodadas, ou caso um dos jogadores ganhe 3 rodadas.
  - - Ao final de cada Rodada, pergunte se o jogador deseja desistir ou continuar o jogo.

````python
class Baralho:
    """Representa um conjunto completo de cartas"""
    
    def __init__(self, cor_do_fundo=0):
        self.cartas = []
        self.cor = cor_do_fundo
        for naipe in range(4):
            for valor in range(1, 14):
                carta = Carta(naipe, valor)
                self.cartas.append(carta)

    def __str__(self):
        cartas = []
        for carta in self.cartas:
            cartas.append(str(carta))
        return '\n'.join(cartas)

    def pop_carta(self):
        return self.cartas.pop()

    def add_carta(self, carta):
        self.cartas.append(carta)

    def embaralhar(self):
        random.shuffle(self.cartas)

    def move_cartas(self, mao, num):
        for i in range(num):
            mao.add_carta(self.pop_carta())

class Carta:
    """Representa uma carta padrão de jogo."""

    naipe_nomes = ['\033[0;34m'+'♣'+'\033[0;0m'+'(Paus)',
                   '\033[1;31m'+'♦'+'\033[0;0m'+'(Ouros)', 
                   '\033[1;31m'+'♥'+'\033[0;0m'+'(Copas)', 
                   '\033[0;34m'+'♠'+'\033[0;0m'+'(Espadas)']
    valor_nomes = [None, '    Ás', '     2', '     3',
                   '     4', '     5', '     6','     7', 
                   '     8', '     9', '    10', 'Valete',
                   '  Dama', '   Rei']

    def __init__(self, naipe=0, valor=2, cor_do_fundo=0):
        self.naipe = naipe
        self.valor = valor
        self.cor = cor_do_fundo
    def __str__(self):
        return '%s de %s' % (Carta.valor_nomes[self.valor],
                                           Carta.naipe_nomes[self.naipe])

    def __lt__(self, other):
        t1 = self.naipe, self.valor
        t2 = other.naipe, other.valor
        return t1 < t2

    def __eq__(self, other):
        t1 = self.naipe, self.valor
        t2 = other.naipe, other.valor
        return t1 == t2

    def __gt__(self, other):
        t1 = self.naipe, self.valor
        t2 = other.naipe, other.valor
        return t1 > t2    

class Mao(Baralho):
    """Representa as cartas na mão do jogador."""

    def __init__(self, label=''):
        self.cartas = []
        self.label = label
    def ordenar(self):
        self.cartas.sort()
````
