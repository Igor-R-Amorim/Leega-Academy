#### Obetendo a media de idade das Mulheres

```python
#Filtrando apenas as mulheres com idade preenchida
filtro1 = df_test['Sex'] == 'female'
filtro2 = df_test['Age'].notna()

# encontrando a quantidade de mulheres sem nulo na idade
qtd_mulheres_sem_nulo = len(df_test.loc[filtro1 & filtro2])

# somando a idade dessas mulheres sem nulo na idade
soma_de_idades_mulheres = df_test.loc[filtro1].loc[filtro2].Age.sum()

# A media nada mais é que a soma da idade dessas mulheres divido 
# por sua quantidade 
# print('%.1f' % (soma_de_idades_mulheres/qtd_mulheres_sem_nulo))
media_de_idades_mulheres = math.ceil(
    soma_de_idades_mulheres/qtd_mulheres_sem_nulo)

print(media_de_idades_mulheres)
```

Observa-se que a media atingida é proxima a média geral, sem uma descrepancia muito grande.

```python
# Vamos substituir agora os campos nulos de idades de mulheres pela media obtida

#Filtrando apenas as mulheres com idade não preenchida
filtro1 = df_test['Sex'] == 'female'
filtro2 = df_test['Age'].isna()

#Listagem dos passageiros onde a idade nao foi prenchida
display(df_test.loc[filtro1 & filtro2].head(3))
print('\n')
display(df_test.loc[filtro1 & filtro2].tail(3))

#Listagem dos indices onde a idade nao foi prenchida
print('\n')
df_test['Age'].loc[filtro1 & filtro2]
```

```python
# Substituindo os nulos pela media de mulheres
df_test.loc[filtro1,'Age'] = df_test.loc[filtro1,'Age'].fillna(media_de_idades_mulheres)
#df_test.loc[filtro1]

# Conferindo a substituição em alguns passageiros
filtro1 = df_test['PassengerId'] == 1302
filtro2 = df_test['PassengerId'] == 914
filtro3 = df_test['PassengerId'] == 1003
display (df_test[filtro1 | filtro2 | filtro3])
```

```python
# A fim de evitar ficar criando e excluindo colunas para executar a conta
# como feita no excel, optou-se por percorrer as linhas somente das 
# colunas desejadas e já verificar as condiçoes da regressao porposta
# e calcula-la retornando apenas previsao do passageiro, se este provavelmente
# vai sobreviver ou não.  

survived = []

for i, infos in df_test[['Sex','Pclass','Age']].iterrows():
  if infos.Sex == 'female':
    Ifem = 1;
  else:
    Ifem = 0;
  if infos.Pclass == 1:
    I1classe = 1
    I2classe = 0
  elif infos.Pclass == 2:
    I1classe = 0
    I2classe = 1
  else:
    I1classe = 0
    I2classe = 0
  idade = float(infos.Age)

  regressao = (-1.33 + (2.55 * Ifem) + (1.27 * I2classe) + (2.58 * I1classe) - 0.04 * idade)
  if regressao >= 1:
    survived.append(3) #Sobrevivera
  else:
    survived.append(2) #Não Sobrevivera

print(len(survived))
print(list(set(survived)))

# Atribuindo a lista criada pela função do for no DF com o nome 'Survived'
# A fim de facilitar a união depois
df_test['Survived'] = survived

# Zerando a variavel apos a atribuição para liberar espaço 
survived = []
```

```python
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pacoty.mplstyle')
#plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')

def grafico_pizza(coluna, titulo='', rotulos=[]):
  dados = coluna.value_counts()
  # Rotulos: 'Legenda: Valor'
  if not bool(rotulos):
    #print("list is empty")
    rotulo_lista = pd.Series(coluna.value_counts().keys().astype(str)).str.cat(
                              list(coluna.value_counts().astype(str)), sep=': ')
  else:
    #print("list is not empty")
    rotulo_lista = pd.Series(rotulos).str.cat(
                              list(coluna.value_counts().astype(str)), sep=': ')
  
  

  sns.color_palette("dark:#5A9_r", as_cmap=True)
  sns.set_style("darkgrid")
  sns.set(rc={'figure.figsize':(14,7)})
  sns.set_palette("viridis")
  plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pacoty.mplstyle')

  f, ax = plt.subplots(figsize=(16,8))
  plt.title(titulo, weight='bold', fontsize = 15)
  plt.pie(dados, labels=rotulo_lista, autopct='%1.2f%%', textprops={'size': 12, 'color':'k'})
  ax.legend(loc='best', frameon=True, fancybox=True, framealpha=0.55, facecolor='white',)
  plt.grid()
  plt.show()

def rotulos_de_barras(x,y): 
  total=0
  y_ajustado=[]

  #calcula o total do eixo y para tirar a percentagem
  for i in y:
    total += float(i)
  
  #Ajuste para barras muito grandes nao ficarem fora do desenho
  cinco_pct = total*0.05
  for i in range(len(y)):
    if y[i] < cinco_pct:
      y_ajustado.append(y[i])
    else:
      y_ajustado.append(y[i]-cinco_pct)
  
  # plota varios textos nas posiçoes x, y ,(string desejada)
  for i in range(len(x)):
      plt.text(i, y_ajustado[i], (str(round(((y[i]/total)*100),1)) + '%\n' + str(y[i])), ha = 'center')
  
  #zera as variaveis para evitar problema no re-uso
  y_ajustado=[]

def grafico_barras(coluna, titulo='', legenda=''):
  
  # dados
  x = list(coluna.value_counts().keys().astype(str))
  y = list(coluna.value_counts())
  if legenda == '':
    label_x = coluna.name
    label_y = 'Quantidade de passageiros com ' + coluna.name
  else:
    label_x = legenda
    label_y = 'Quantidade de passageiros com ' + legenda
  
  # ordena os índices em vez dos elementos em si
  indices = list(range(len(x)))
  indices.sort(key=lambda i: x[i]) # ordene os índices com relação ao seu respectivo valor em x
  # cria as listas baseado na ordem dos índices
  eixo_x = [x[i] for i in indices]
  eixo_y = [y[i] for i in indices]
  
  # plot
  bar_labels = eixo_x
  bar_colors = ['#D94F94','#18c0c4','#7A76C2','#7fcdbb','#ef8dac', '#66E9EC', '#bcbddc']
  fig, ax = plt.subplots(figsize=(14, 7))
  ax.bar(eixo_x, eixo_y, width=0.7, edgecolor='white', linewidth=0.9, alpha=0.9, label=bar_labels, color=bar_colors)
  rotulos_de_barras(eixo_x, eixo_y)
  ax.set_xlabel(label_x)
  ax.set_ylabel(label_y)
  ax.set_title(titulo, weight='bold', fontsize = 15)
  #ax.legend([label_y], loc='best', frameon=True, fancybox=True, framealpha=0.55, facecolor='white', prop={'size':12}, ncol=3)
  
  plt.show()

def grafico_barras_empilhadas (df, coluna_1, coluna_2, titulo='', rotulos=[]):

  #dados
  try:
    df['Pclass'].replace(1, '1ª Classe', inplace=True)
    df['Pclass'].replace(2, '2ª Classe', inplace=True)
    df['Pclass'].replace(3, '3ª Classe', inplace=True)
  except:
    pass
  
  try:
    df['Sex'].replace('female', 'Feminino', inplace=True)
    df['Sex'].replace('male', 'Masculino', inplace=True)
  except:
    pass
  
  try:
    df['Survived'].replace(0, 'Não Sobreviveu', inplace=True)
    df['Survived'].replace(1, 'Sobreviveu', inplace=True)
    df['Survived'].replace(2, 'Não Sobreviverá', inplace=True)
    df['Survived'].replace(3, 'Sobreviverá', inplace=True)
  except:
    pass
  
  label_1 = coluna_1.name
  label_2 = coluna_2.name
  df_agrupado = df.groupby([label_1, label_2]).size().unstack()
  
  # Rotulos: ['label_1','label_2']
  if not bool(rotulos):
    #print("list is empty")
    label_x = label_1
    label_y = 'Quantidade de ' + label_2
  else:
    label_1 = rotulos[0]
    label_2 = rotulos[1]
    label_x = label_1
    label_y = 'Quantidade de ' + label_2  
  
  # plot de um gráfico de barras empilhadas utilizando o caminho do Pandas (df.plot)
  ax = df_agrupado.plot.bar(stacked=True, color=('#D94F94','#18c0c4','#7A76C2','#7fcdbb','#ef8dac', '#66E9EC', '#bcbddc'), figsize=(14, 7))
  ax.set_xlabel(label_x)
  ax.set_ylabel(label_y)
  ax.set_title(titulo, weight='bold', fontsize = 15)
  ax.set_title(titulo, weight='bold', fontsize = 15)
  ax.legend(list(df_agrupado.columns), loc='best', frameon=True, fancybox=True, framealpha=0.55, facecolor='white', prop={'size':12})
  # Percorre a lista de plotagens no eixo ax
  # criando uma nova lista de rotulos para cada
  # item desenhado 
  for c in ax.containers:
    rotulos = [a if a else "" for a in c.datavalues]
    pct=[]
    pula_linha = []
    # Criei essa parte apenas para conter a porcentagem
    for i in range(len(rotulos)):
      pct.append( round( ((rotulos[i]/len(df))*100), 1) )
    rotulo_pct = pd.Series(pct).astype(str).str.cat(pd.Series(rotulos).astype(str), sep='%  ')
    # imprimindo os rotulos
    ax.bar_label(c, labels=rotulo_pct, label_type="center")
  plt.xticks(rotation=0)
  plt.show()

def grafico_histograma (df, coluna, titulo='', legenda=''):
  # dados
  x = list(coluna.value_counts().keys().astype(str))
  y = list(coluna.value_counts())
  if legenda == '':
    label_x = coluna.name
    label_y = 'Quantidade de passageiros com faixa de ' + coluna.name
  else:
    label_x = legenda
    label_y = 'Quantidade de passageiros com faixa de ' + legenda  

  
  # ordena os índices em vez dos elementos em si
  indices = list(range(len(x)))
  indices.sort(key=lambda i: x[i]) # ordene os índices com relação ao seu respectivo valor em x
  # cria as listas baseado na ordem dos índices
  eixo_x = [x[i] for i in indices]
  eixo_y = [y[i] for i in indices]
  
  # configurando o estilo e a folha no seaborn
  sns.set(rc={'figure.figsize':(14,7)})
  sns.set_palette("viridis")
  plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pacoty.mplstyle')

  # plot de um gráfico de HISTOGRAMA utilizando o caminho do Seaborn (sns.plot)
  ax = sns.histplot(data=df, x=coluna, binwidth=10, kde=True, label='Age', palette='viridis')
  plt.xlabel(label_x)
  plt.ylabel(label_y)
  plt.title(titulo+'\n', weight='bold', fontsize = 15)
  plt.legend(['Densidade',label_x], loc='best', frameon=True, fancybox=True, framealpha=0.55, facecolor='white', prop={'size':12})
  ax.bar_label(ax.containers[0], label_type='center',) 
  #plt.plot(eixo_x, eixo_y)
  plt.xticks(list(range(0, len(x)+10, 10)), labels=list(range(0, len(x)+10, 10)),)
  plt.show()

```

```python
def leiaInt(msg = ''):  # inserido após decidir colocar a opção desejada pelo cliente
    while True:  # lendo até ser correto
        try:
            n = int(input(msg))
        except (ValueError, TypeError):
            print(f'\033[31m', 'ERRO! Por favor, digite uma opção válida válido.', '\033[0;0m')
            continue
        else:
            return n
            
def menu(lista, titulo):
  opc = 0
  while True:
    if (opc-1 >= len(lista) or opc <= 0):
      print('-'*50)
      print(titulo.center(50))  # certo da quantidade de caracteres
      print('-'*50)

      # para não ficar como lista, criar um contador:
      for i, item in enumerate(lista, start=1):
          print(f'\033[34m', i, '\033[0;0m', '-','\033[1;36m', item,'\033[0;0m')
      
      print() 
      opc = leiaInt('Sua Opção: ')
      clear_output(wait=True)
    else:
      print(f'A opção escolhida foi: {opc} - {lista[opc-1]}')
      break
  return opc
```
